```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta,date
import re
```

```python
path = r"C:\Users\NT550-052\OneDrive\바탕 화면\ASAC\프로젝트\ML\web_crawling\merge_dataset\raw_merge.csv"

data = pd.read_csv(path,sep="\t")

data
```

```python
#  days_after_release
today = "2023-05-19"
today = datetime.strptime(today,'%Y-%m-%d')
data['new_date'] = pd.to_datetime(data['new_date'])
data["days_after_release"] = data["new_date"].apply(lambda x: (today - x).days)
```

```python
# compare is free 

for idx, elem in enumerate(data["is_free"]):
    if elem == 0.0:
        data["is_free"][idx] = 0
    else:
        if data["price"][idx] != 0:
            data["is_free"][idx] = 0
        else:
            data["is_free"][idx] = 1
sum(data["is_free"].isnull())

data["is_free"] = data["is_free"].astype("int")
```

```python
# 24_Hour_Peak, All_time_peak nan replace
# 24_Hour_Peak -> 0
# All_time_peak -> average_forever value

for i in range(len(data)):
    x = data["24_Hour_Peak"][i]
    y = data["All_time_peak"][i]
    
    if x == "Invalid":
        data["24_Hour_Peak"][i] = 0
    if y == "Invalid":
        data["All_time_peak"][i] = data["average_forever"][i]
```

```python
# windows, mac, linux, platforms_num nan -> 0
data.loc[:,["windows","mac","linux","platforms_num"]] = data.loc[:,["windows","mac","linux","platforms_num"]].fillna(0)
```

```python
data.loc[:,["windows","mac","linux","platforms_num"]]
```

```python
# genre None -> "N"
data["genre"] = data["genre"].replace("None", "N")
# Review nan -> replace "No user reviews"
data["Review"] = data["Review"].fillna("No user reviews")

# reorder column
data = data[['appid', '24_Hour_Peak', 'All_time_peak', 'positive', 'negative',
       'average_forever', 'average_2weeks', 'median_forever', 'median_2weeks',
       'price', 'initialprice', 'genre', 'num_lang', 'required_age', 'is_free',
       'controller_support', 'dlc', 'metacritic', 'metacritic_tf', 'windows',
       'mac', 'linux', 'platforms_num', 'recommendations', 'achievements',
       'new_date', 'days_after_release','Review']]
```
